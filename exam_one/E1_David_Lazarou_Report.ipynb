The expected return of a portfolio is given by:

$E(R_p) = \sum_{i=1}^{n} w_i E(R_i)$

The variance of a portfolio is:

$$
\sigma_p^2 = \sum_{i=1}^{n} w_i^2 \sigma_i^2 + \sum_{i=1}^{n} \sum_{j=1, j \neq i}^{n} w_i w_j \sigma_{i,j}
$$


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample stock returns
returns = np.array([
    [0.12, 0.18, 0.10],
    [0.10, 0.15, 0.12],
    [0.09, 0.14, 0.11]
])

# Compute covariance matrix
cov_matrix = np.cov(returns.T)

# Display covariance matrix
pd.DataFrame(cov_matrix, columns=['Stock A', 'Stock B', 'Stock C'], index=['Stock A', 'Stock B', 'Stock C'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample Efficient Frontier
returns = np.linspace(0.02, 0.20, 100)
volatility = np.sqrt(returns)

plt.figure(figsize=(8,5))
plt.plot(volatility, returns, label="Efficient Frontier")
plt.xlabel("Portfolio Volatility")
plt.ylabel("Expected Return")
plt.title("Efficient Frontier for Portfolio Optimization")
plt.legend()
plt.show()
```
